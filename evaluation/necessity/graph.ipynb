```python
import matplotlib.pyplot as plt
import numpy as np

# Data
methods = ['AuditGPT', 'W.O. RE', 'W.O. S', 'W.O. OS', 'W.O. CM', 'W.O. CS']
TP = [139, 70, 83, 120, 131, 71]
FP = [3, 96, 240, 59, 182, 132]

# The x position of bars
# x = np.arange(len(methods))
x = np.array([0,2,3,4,5,6])

# Width of a bar
width = 0.37

# Create the figure and an axes set
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting
ax.bar(x - width*0.5, TP, width, label='TP', color='none', edgecolor='green', linewidth=1, hatch='...')  # Small dots pattern for TP
ax.bar(x + width*0.5, FP, width, label='FP', color='none', edgecolor='red', linewidth=1, hatch='///')  # Lines pattern for FP

# X axis configuration
ax.set_xticks(x)
ax.set_xticklabels(methods, fontsize=20)
ax.grid(False)
ax.tick_params(axis='y', labelsize=20)

ax.bar_label(ax.containers[0], color='green', fontsize=19)
ax.bar_label(ax.containers[1], color='red', fontsize=19)

ax.legend(prop={'size':20})

# Show the plot
plt.tight_layout()
plt.show()


```
